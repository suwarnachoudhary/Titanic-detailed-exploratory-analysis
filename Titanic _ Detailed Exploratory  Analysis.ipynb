```python
# Function to extract th title from the name
def GetTitle(name):
    title_group = {'mr':'Mr',
                   'mrs': 'Mrs',
                   'miss':'Miss',
                   'master':'Master',
                   'don':'Sir',
                   'rev':'Sir',
                   'dr':'Officer',
                   'mme':'Mrs',
                   'ms':'Mrs',
                   'major':'Officer',
                   'mlle':'Miss',
                   'lady':'Lady',
                   'sir':'Sir',
                   'col':'Officer',
                   'capt':'Officer',
                   'the countess':'Lady',
                   'jonkheer':'Sir',
                   'dona':'Lady'
                  }
    first_name_with_title=name.split(',')[1]
    title=first_name_with_title.split('.')[0]
    title=title.strip().lower()
    return title_group[title]
```

```python
df['Title'] = df.Name.map(lambda x: GetTitle(x))
```
